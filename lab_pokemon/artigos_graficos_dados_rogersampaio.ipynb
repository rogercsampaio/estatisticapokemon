```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Gráfico de linhas

```python
vendas_dados = {'01/2024':120.500,'02/2024':130.600,
             '03/2024':100.231,'04/2024':101.5600,
             '05/2024':102.500,'06/2024':105.000,
             '07/2024':145.030,'08/2024':151.903,
             '09/2024':160.000,'10/2024':108.230,
             '11/2024':161.921,'12/2024':101.00}
```

```python
df_vendas = pd.DataFrame(list(vendas_dados.items()), columns=['Mês', 'Vendas'])
print(df_vendas)
```

```python
# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(df_vendas['Mês'], df_vendas['Vendas'], marker='o', linestyle='-', color='b', label='Vendas')

# Configurando título e rótulos
plt.title('Vendas Mensais em 2024', fontsize=14)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Vendas (R$)', fontsize=12)

# Exibindo as vendas no gráfico
plt.xticks(rotation=45)  # Rotaciona as etiquetas do eixo x para melhor visualização
plt.grid(True)

# Exibindo o gráfico
plt.tight_layout()
plt.legend()
plt.show()
```

```python
df_vendas_2023_2024 = pd.DataFrame(list(vendas_dados_2023_2024.items()), columns=['Mês', 'Vendas'])
print(df_vendas_2023_2024)
```

```python
df_vendas_2023_2024['Mês-Isolado'] = df_vendas_2023_2024['Mês'].str.split('/').str[0]
df_vendas_2023_2024
```

```python
df_vendas_2023_2024_agrupado = df_vendas_2023_2024.groupby('Mês-Isolado').sum()
df_vendas_2023_2024_agrupado.drop('Mês', axis=1, inplace=True)
df_vendas_2023_2024_agrupado
```

```python
# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(df_vendas_2023_2024_agrupado.index, df_vendas_2023_2024_agrupado['Vendas'],
         marker='o', linestyle='-', color='b', label='Vendas')

# Configurando título e rótulos
plt.title('Vendas Mensais em 2023 e 2024 (somatório)', fontsize=14)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Vendas (R$)', fontsize=12)

# Exibindo as vendas no gráfico
plt.xticks(rotation=45)  # Rotaciona as etiquetas do eixo x para melhor visualização
plt.grid(True)

# Exibindo o gráfico
plt.tight_layout()
plt.legend()
plt.show()
```

```python
import pandas as pd

# Dados fornecidos
dados = [
    ['01/2023', 'Adidas', 42.600],
    ['01/2023', 'Nike', 60.200],
    ['01/2023', 'Reebok', 38.700],
    ['02/2023', 'Adidas', 56.100],
    ['02/2023', 'Nike', 100.000],
    ['02/2023', 'Reebok', 34.200],
    ['03/2023', 'Adidas', 45.200],
    ['03/2023', 'Nike', 50.000],
    ['03/2023', 'Reebok', 5.091]
]

# Criar o DataFrame
df_vendas_jan_marco_2023 = pd.DataFrame(dados, columns=['mes', 'marca', 'vendas'])

# Exibir o DataFrame
print(df_vendas_jan_marco_2023)
```

```python
# Uma das formas de agrupar é pivotando o dataframe (alterando a estrutura). Poderíamos ter o mesmo resultado com group_by
df_pivot = df_vendas_jan_marco_2023.pivot_table(index='mes', columns='marca', values='vendas', aggfunc='sum')

# Gerar o gráfico de barras agrupadas
df_pivot.plot(kind='bar', figsize=(10, 6))

# Ajustar título e rótulos
plt.title('Vendas por Marca e Mês (Jan-Mar 2023)')
plt.xlabel('Mês')
plt.ylabel('Vendas')

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
