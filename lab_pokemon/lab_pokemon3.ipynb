```python
# Importação de pacotes e bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import statistics as st
import math as mt
```

```python
# Carregando a base de dados
pokemons = pd.read_csv("Pokemon.csv")
print(f"Base de dados carregada com sucesso. Dimensões {pokemons.shape[0]} linhas e {pokemons.shape[1]} colunas")
print(f"Nome de colunas {pokemons.columns}")
```

```python
# Visualizando 3 primeiros pókemons, alguns valores
pokemons[['Name','Speed','HP']].head(3)
```

```python
plt.xlabel('Speed(Velocidade)')
plt.ylabel('HP(Saúde)')
plt.title('Análise de Speed vs HP')
plt.scatter(pokemons['Attack'], pokemons['Sp. Atk'])
plt.plot([0,300], [0,300], 'k-')
plt.show()
```

```python
print("Coeficiente de correlação")
pokemons['Speed'].corr(pokemons['HP'])
```

```python
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Análise de Attack vs Defense')
plt.scatter(pokemons['Attack'], pokemons['Defense'])
plt.plot([0,300], [0,300], 'k-')
plt.show()
```

```python
print("Coeficiente de correlação")
pokemons['Attack'].corr(pokemons['Defense'])
```
